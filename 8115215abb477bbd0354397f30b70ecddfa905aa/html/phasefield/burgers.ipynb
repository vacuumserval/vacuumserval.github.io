### Burgers取向关系

在钛合金中，最常见的是$\beta$相和$\alpha$相之间的转变。$\beta$相具有立方结构，而$\alpha$相具有密排六方结构。在$\beta$相转变为$\alpha$相的过程中，存在一种特定的取向关系，即Burgers取向关系，这种关系的具体表达式是

$$
\begin{align}
x_o\parallel [100] \parallel [1\bar{1}0]_\beta &\parallel [0001]_\alpha \\
y_o\parallel [010] \parallel [110]_\beta &\parallel [10\bar{1}0]_\alpha \\
z_o\parallel [001] \parallel [001]_\beta &\parallel [\bar{1}2\bar{1}0]_\alpha
\end{align}
$$

这意味着在$\beta$相晶粒内，可能会出现12个不同取向的$\alpha$相变体。

<img src="https://www.mdpi.com/materials/materials-15-05325/article_deploy/html/images/materials-15-05325-g001.png" width="400" alt="p1">

这种取向关系的存在对钛合金的力学性能有重要影响。例如，它影响了合金的强度、塑性和疲劳性能。因此，理解并控制这种变体选择过程是提高钛合金性能的关键。目前，关于钛合金中Burgers取向关系的研究仍在进行中，特别是关于如何在相变过程中控制和减弱变体选择的发生，这被认为是进一步提高钛合金性能的潜在方法之一。

<img src="./image/burgersog.png" width="400" alt="p1"> [2]

具体到原子来看，就是要求BCC结构（灰色）原子，和HCP结构（蓝色）原子在三个正交方向上重合，转移矩阵为$F_{tot}$，BCC到HCP的转换的数学表示为$\mathbf{r}_\alpha=F_{tot}\mathbf{r}_\beta$。这需要通过压缩/拉伸变形(e, $F_1$)+旋转(f, $R$)来实现。

* 先进行压缩/拉伸：

   * 沿着$[010]_\beta$方向，即$[\bar{1}2\bar{1}0]_\alpha$方向，BCC和HCP的两套格子的原子最近临间距分别为$a_\beta \Leftrightarrow a_\alpha$;
   * 沿着$[\bar{1}01]_\beta$方向，即$[\bar{1}0\bar{1}0]_\alpha$方向，BCC和HCP的两套格子的原子最近临间距分别为$\sqrt{2}a_\beta \Leftrightarrow \sqrt{3}a_\alpha$;
   * 沿着$[101]_\beta$方向，即$[0001]_\alpha$方向，BCC和HCP的两套格子的原子最近临间距分别为$\sqrt{2}a_\beta \Leftrightarrow c_\alpha$;

   因此压缩/拉伸变形转移矩阵$F_1$为

   $$
   \begin{align}
   F_1 =
   \begin{pmatrix}
   a_\alpha / a_\beta & 0 & 0 \\
   0 & \sqrt{3}a_\alpha/\sqrt{2}a_\beta & 0 \\
   0 & 0 & c_\alpha/\sqrt{2}a_\beta
   \end{pmatrix}
   \end{align}
   $$

* 再进行$\frac{70.5\degree-60\degree}{2}\approx 5.25\degree$的旋转，旋转矩阵$R$为

   $$
   \begin{align}
   R =
   \begin{pmatrix}
   cos(5.25\degree) & sin(5.25\degree) & 0 \\
   -sin(5.25\degree) & cos(5.25\degree) & 0 \\
   0 & 0 & 1
   \end{pmatrix}
   \end{align}
   $$

转移矩阵为$F_{tot}=RF_1$，全部12种变体的转移矩阵可以通过对称性获得。

钛合金中的马氏体相$\alpha'$也是HCP结构的，马氏体相变通常是从高温的$\beta$相（体心立方结构，BCC）通过快速冷却，转变为$\alpha'$马氏体相（密排六方结构，HCP）的过程。马氏体相变不涉及原子扩散，而是通过原子重新排列来形成。*Theoretically, unlike diffusion transformations, martensitic transformations (also known as displacive transformations) involve short-range atomic displacements, which result in a change of crystal structure with minimum transformation elastic strain energy. Based on the Burgers orientation correspondence, for the martensitic transformation of Ti-6Al-4V alloy, the lattice orthohexagonal reference system between BCC and HCP cells is shown in Figure 1* [1] - 理论上，与扩散型相变不同，马氏体相变涉及短程原子位移，这导致晶体结构发生变化，同时使变形弹性应变能最小化。对于马氏体相变，论文[1]指出*Theoretically, the atomic shuffle is necessary for phase transformation between BCC and HCP crystallographic structures through dislocation glide [52–54]. Therefore, in the current work, without considering the internal shuffle, only six correspondence variants from a single parent beta grain are produced (in order to distinguish from the 12 crystallographic variants in which the dislocation glide is under consideration, the phrase ‘correspondence variant’ is used, and ‘variant’ refers to the correspondence variant without specifying the ‘crystallographic variants’).* - 理论上，原子重排对于通过位错滑移在BCC（体心立方）和HCP（六方密排）晶体结构之间进行相变是必要的。因此，在当前工作中，没有考虑内部重排，仅从单一母β晶粒产生了6个对应变体（为了与考虑位错滑移时的12个晶体学变体相区分，这里使用了“对应变体”这个术语，而“变体”指的是不考虑“晶体学变体”的对应变体）。我的理解（不一定对）这就意味着，在论文[1]马氏体的建模中，就没有滑移引起的那个旋转矩阵，而是只有压缩/拉伸矩阵了。另一方面，[1]与[2]选取了不同的正交坐标系方向，为方便后续编程参考，在此采取[1]的坐标选取法和对应变体指定法，即

* 变形矩阵$U_o$

$$
\begin{align}
\mathbf{U}_o =
\begin{pmatrix}
\gamma=c_\alpha/\sqrt{2}a_\beta & 0 & 0 \\
0 & \alpha=\sqrt{3}a_\alpha/\sqrt{2}a_\beta & 0 \\
0 & 0 & \beta=a_\alpha / a_\beta
\end{pmatrix}
\end{align}
$$

注意到不同成分和热处理制度钛合金$\alpha$和$\beta$相的晶格常数不同，这就使得模型具有了建模不同钛合金的能力。*Applying the Cauchy–Born Hypothesis that links the lattice to the continuum, the transformation matrix of variant 1 can be identified as* [3] - 根据Cauchy–Born假设，将晶格与连续体相联系，变体1的变换矩阵可以被定义为

$$
\begin{align}
\mathbf{U}_p =
\begin{pmatrix}
\frac{\alpha+\gamma}{2} & \frac{\alpha-\gamma}{2} & 0 \\
\frac{\alpha-\gamma}{2} & \frac{\alpha+\gamma}{2} & 0 \\
0 & 0 & \beta
\end{pmatrix}
\end{align}
$$

对于论文[1]的钛合金，晶格常数$a_\beta=3.3065\AA$，$a_\alpha=2.9511\AA$，$c_\alpha=4.6843\AA$。

```python
import numpy as np
def Ucal(a_beta, a_alpha, c_alpha): # 计算变换矩阵(3D,3D,6variant)
    U = np.zeros((3,3,6)) # 变换矩阵(3D,3D,6variant)
    gamma = c_alpha / (np.sqrt(2)*a_beta)
    alpha = (np.sqrt(3)*a_alpha)/(np.sqrt(2)*a_beta)
    beta = a_alpha / a_beta
    u11 = 0.5*(alpha+gamma)
    u12 = 0.5*(alpha-gamma)
    u33 = beta
    U[:,:,5] = np.array([[u11, u12, 0],\
                   [u12, u11, 0],\
                    [0, 0, u33]])
    U[:,:,4] = np.array([[u11, -u12, 0],\
                   [-u12, u11, 0],\
                    [0, 0, u33]])
    U[:,:,3] = np.array([[u11, 0, u12],\
                   [0, u33, 0],\
                    [u12, 0, u11]])
    U[:,:,2] = np.array([[u11, 0, -u12],\
                   [0, u33, 0],\
                    [-u12, 0, u11]])
    U[:,:,1] = np.array([[u33, 0, 0],\
                   [0, u11, u12],\
                    [0, u12, u11]])
    U[:,:,0] = np.array([[u33, 0, 0],\
                   [0, u11, -u12],\
                    [0, -u12, u11]])
    return(U)
ab = 3.3065
aa = 2.9511
ca = 4.6843
```

## 马氏体相变(Martensite Transformation, MT)相场模型

*During martensitic transformation, the change of lattice structure leads to transformation deformation. The SFTS can be calculated with following equation* [1] - 在马氏体相变过程中，晶格结构的变化导致相变变形。形状变化引起的特征应变（SFTS）可以通过以下方程计算得出。

$$
\begin{align}
\varepsilon_{ij}^{00}(p)=\frac{\mathbf{U}_p'\mathbf{U}_p-\mathbf{I}}{2}
\end{align}
$$

```python
import numpy as np

def e2cal(U): # 计算特征应变(3D,3D,6variant)
    e2 = np.zeros((3,3,6)) # 特征应变(3D,3D,6variant)
    for p in range(6):
        e2[:,:,p] = (np.matmul(np.transpose(U[:,:,p]), U[:,:,p])-np.eye(3))/2
    return(e2)

ab = 3.3065
aa = 2.9511
ca = 4.6843

U = Ucal(a_beta=ab, a_alpha=aa, c_alpha=ca)
e2 = e2cal(U=U)
for p in range(6):
    print('U_',str(p+1),'=', U[:,:,p])
    print('e_',str(p+1),'=', e2[:,:,p])
```

有些文献（如[4]），应变张量的求法略有不同。

$$
\begin{align}
\varepsilon_{ij}^{00}(p)=\frac{\mathbf{U}_p'+\mathbf{U}_p}{2}-\mathbf{I}
\end{align}
$$
无论哪种定义法，总之都是给应变增加了一定的各项异性。

系统的自由能泛函[1]建模为：

$$
\begin{align}
F = \int\left(F^{ch}+F^{grad}\right)d^3\mathbf{r} + F^{el}
\end{align}
$$

其中，化学能泛函$F^{ch}$是Landau多项式形式的：

$$
\begin{align}
F^{ch} = \frac{1}{V_m}\left[\frac{A}{2}\sum_{p=1}^6\eta_p^2-\frac{B}{3}\sum_{p=1}\eta_p^3+\frac{C}{4}\left(\sum_{p=1}^6\eta_p^2\right)^2\right]
\end{align}
$$

界面能泛函$F^{grad}$是梯度场形式的（这里[1]论文写的不甚清楚，代码上反推是这样）：
$$
\begin{align}
F^{grad} = \frac{1}{V_m}\left[\sum_{p=1}^6\frac{\kappa}{2}\left(\nabla\eta_p\right)^2\right]
\end{align}
$$

弹性应变能$F^{el}$可以根据KS理论进行公式化表达：
$$
\begin{align}
F^{el} = \frac{1}{2}\sum_{p=1}^{6}\oint\frac{d^3\mathbf{k}}{(2\pi)^3}B_{pq}(\mathbf{n})\{\hat{\eta}_p\}_k\{\hat{\eta}_q\}_k^*
\end{align}
$$

其中积分是在倒易空间中进行的，$\mathbf{k}$是倒易空间中的一个矢量，$\mathbf{n}=\mathbf{k}/|k|$是倒易空间中的单位矢量。注意，积分中要排除$k = 0$的情况。$\{\hat{\eta}_p\}_k$是$\eta(\mathbf{r})$的傅里叶变换，其中$\mathbf{r}$是实空间中的矢量。星号表示复共轭。考虑一个具有限定（即固定）外部边界（零应变条件）的系统， $B_{pq}(\mathbf{n})$的形式为：
$$
\begin{align}
B_{pq}(\mathbf{n}) = 
\begin{cases}
C_{ijkl}\varepsilon_{ij}(p)\varepsilon_{kl}(q) \;\; &\; \mathbf{n}=0 \\
C_{ijkl}\varepsilon_{ij}(p)\varepsilon_{kl}(q)-n_i\sigma_{ij}^p\Omega_{jk}\sigma_{kl}^qn_l \;\; &\; \mathbf{n}\neq 0 
\end{cases} 
\end{align}
$$
其中$C_{ijkl}$是弹性刚度张量，在均匀模拟系统中，假设其$\alpha$相和$\beta$相的组分相同，并且是各向同性的，大部分文献采用$\alpha$相的$C_{ijkl}$。在Voigt表示法中，四阶弹性张量$C_{ijkl}$的独立分量被映射到二阶弹性张量$C_{ij}$的6个独立分量，如下所示：
$$
\begin{align}
C_{11} &\Leftrightarrow C_{1111} \\
C_{22} &\Leftrightarrow C_{2222} \\
C_{33} &\Leftrightarrow C_{3333} \\
C_{12}=C_{21} &\Leftrightarrow C_{1122} \\
C_{13}=C_{31} &\Leftrightarrow C_{1133} \\
C_{23}=C_{32} &\Leftrightarrow C_{2233}
\end{align}
$$
此外，还有三个剪切模量，它们是：
$$
\begin{align}
C_{44} &\Leftrightarrow C_{1212}\; \text{or} \; C_{2323} \\
C_{55} &\Leftrightarrow C_{1313} \\
C_{66} &\Leftrightarrow C_{1212}=C_{1313}=C_{2323}
\end{align}
$$

论文[1]中的弹性常数$C_{11}=97.7GPa$，$C_{12}=82.7GPa$，$C_{44}=7.5GPa$，相场方法使用时应除以$\frac{\Delta G_m}{V_{mol}}=\frac{3.1375e02 J/mol}{1e-5 m^3/mol}$以无量纲化。
