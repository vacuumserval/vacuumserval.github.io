# Phase Field Crystal (2D)

## Mode Expansion

### Triangular (hexagonal) lattice

#### the lattice vectors

* the lattice vectors:

$$
\begin{align}
\begin{cases}
a_1 &= \left(-\frac{2a}{\sqrt{3}}, 0\right) \nonumber \\
a_2 &= \left(-\frac{a}{\sqrt{3}}, a\right) \nonumber
\end{cases}
\end{align}
$$

* the reciprocal lattice vectors (RLVs):

$$
\begin{align}
\begin{cases}
b_1 &= \left(-\frac{\sqrt{3}\pi}{a}, -\frac{\pi}{a}\right) \nonumber \\
b_2 &= \left(0, \frac{2\pi}{a}\right) \nonumber
\end{cases}
\end{align}
$$

the magnitude of the principal RLVs is $\frac{2\pi}{a}$

#### the mode expansion

The density $nlst$ of the crystalline state can be expressed in terms of the reciprocal lattice vectors (RLVs) $\mathbf{k}$'s

$$
\begin{align}
nlst(r) = n_0+\sum_\mathbf{k}A_ke^{i\mathbf{k}\cdot\mathbf{r}}
\end{align}
$$

* 1st mode

The 1st mode consists of $(1,0), (0,1), (-1,-1)$:

$$
\begin{align}
\sum_\mathbf{k_1} A_{1}e^{i\mathbf{k_1} \cdot\mathbf{r}} =A_1\left[ 2 \cos \left(\frac{\pi  \sqrt{3} x}{a}+\frac{\pi  y}{a}\right)+2 \cos \left(\frac{\sqrt{3} \pi  x}{a}-\frac{\pi  y}{a}\right)+2 \cos \left(\frac{2 \pi  y}{a}\right)\right]
\end{align}
$$

* 2nd mode

The 2nd mode consists of $(-1,-2), (-2,-1), (-1,1)$:

$$
\begin{align}
\sum_\mathbf{k_2} A_{2}e^{i\mathbf{k_2} \cdot\mathbf{r}} = A_2\left[2 \cos \left(\frac{\pi  \sqrt{3} x}{a}+\frac{3 \pi  y}{a}\right)+2 \cos \left(\frac{\sqrt{3} \pi  x}{a}-\frac{3 \pi  y}{a}\right)+2 \cos \left(\frac{2 \sqrt{3} \pi  x}{a}\right)\right] 
\end{align}
$$

So the density $nlst$ is

$$
\begin{align}
n_t(r) &= n_0+\sum_\mathbf{k}A_ke^{i\mathbf{k}\cdot\mathbf{r}} \approx n_0+A_{1}\sum_\mathbf{k_1} e^{i\mathbf{k_1} \cdot\mathbf{r}}+ A_{2}\sum_\mathbf{k_2}e^{i\mathbf{k_2} \cdot\mathbf{r}}\nonumber \\
&=n_0+A_1\left[ 2 \cos \left(\frac{\pi  \sqrt{3} x}{a}+\frac{\pi  y}{a}\right)+2 \cos \left(\frac{\sqrt{3} \pi  x}{a}-\frac{\pi  y}{a}\right)+2 \cos \left(\frac{2 \pi  y}{a}\right)\right] \nonumber \\
&+A_2\left[2 \cos \left(\frac{\pi  \sqrt{3} x}{a}+\frac{3 \pi  y}{a}\right)+2 \cos \left(\frac{\sqrt{3} \pi  x}{a}-\frac{3 \pi  y}{a}\right)+2 \cos \left(\frac{2 \sqrt{3} \pi  x}{a}\right)\right] 
\end{align}
$$

in PFC implementations, $A_2$ is set to be $0$.

```python
import numpy as np
import matplotlib.pyplot as plt
def modeexptr(n0, a, A1, A2, x, y):
    m1 = A1*(2*np.cos(np.pi*np.sqrt(3)*x/a+np.pi*y/a)+2*np.cos(np.pi*np.sqrt(3)*x/a-np.pi*y/a)+2*np.cos(2*np.pi*y/a))
    m2 = A2*(2*np.cos(np.pi*np.sqrt(3)*x/a+3*np.pi*y/a)+2*np.cos(np.pi*np.sqrt(3)*x/a-3*np.pi*y/a)+2*np.cos(2*np.sqrt(3)*np.pi*x/a))
    n = n0 + m1 + m2
    return(n)
x = np.arange(-3, 3.1, 0.1)
y = np.arange(-4, 4.1, 0.1)
X, Y = np.meshgrid(x, y, indexing='ij')
ntr = modeexptr(0, 1, 1, 0, X, Y)
plt.imshow(ntr)
```

### Square lattice

#### the lattice vectors

* the lattice vectors:

$$
\begin{align}
\begin{cases}
a_1 &= \left(a, 0\right) \nonumber \\
a_2 &= \left(0, a\right) \nonumber
\end{cases}
\end{align}
$$

* the reciprocal lattice vectors (RLVs):

$$
\begin{align}
\begin{cases}
b_1 &= \left(\frac{2\pi}{a}, 0\right) \nonumber \\
b_2 &= \left(0, \frac{2\pi}{a}\right) \nonumber
\end{cases}
\end{align}
$$

the magnitude of the principal RLVs is $\frac{2\pi}{a}$

#### the mode expansion

The density $n$ of the crystalline state can be expressed in terms of the reciprocal lattice vectors (RLVs) $\mathbf{k}$'s

$$
\begin{align}
n(r) = n_0+\sum_\mathbf{k}A_ke^{i\mathbf{k}\cdot\mathbf{r}}
\end{align}
$$

* 1st mode

The 1st mode consists of $(1,0), (0,1)$:

$$
\begin{align}
\sum_\mathbf{k_1} A_{1}e^{i\mathbf{k_1} \cdot\mathbf{r}} =A_1\left[2 \cos \left(\frac{2 \pi  x}{a}\right)+2 \cos \left(\frac{2 \pi  y}{a}\right)\right]
\end{align}
$$

* 2nd mode

The 2nd mode consists of $(1,1), (1,-1)$:

$$
\begin{align}
\sum_\mathbf{k_2} A_{2}e^{i\mathbf{k_2} \cdot\mathbf{r}} = A_2\left[2 \cos \left(\frac{2 \pi  x}{a}+\frac{2 \pi  y}{a}\right)+2 \cos \left(\frac{2 \pi  x}{a}-\frac{2 \pi  y}{a}\right)\right] 
\end{align}
$$

So the density $nlst$ is

$$
\begin{align}
n_s(r) &= n_0+\sum_\mathbf{k}A_ke^{i\mathbf{k}\cdot\mathbf{r}} \approx n_0+A_{1}\sum_\mathbf{k_1} e^{i\mathbf{k_1} \cdot\mathbf{r}}+ A_{2}\sum_\mathbf{k_2}e^{i\mathbf{k_2} \cdot\mathbf{r}}\nonumber \\
&=n_0+A_1\left[2 \cos \left(\frac{2 \pi  x}{a}\right)+2 \cos \left(\frac{2 \pi  y}{a}\right)\right]+A_2\left[2 \cos \left(\frac{2 \pi  x}{a}+\frac{2 \pi  y}{a}\right)+2 \cos \left(\frac{2 \pi  x}{a}-\frac{2 \pi  y}{a}\right)\right] 
\end{align}
$$

in PFC implementations, $A_2\neq 0$, the exact value need to be determined through further calculation.

```python
import numpy as np
import matplotlib.pyplot as plt
def modeexpsq(n0, a, A1, A2, x, y):
    m1 = A1*(2*np.cos(2*np.pi*x/a)+2*np.cos(2*np.pi*y/a))
    m2 = A2*(2*np.cos(2*np.pi*x/a+2*np.pi*y/a)+2*np.cos(2*np.pi*x/a-2*np.pi*y/a))
    n = n0 + m1 + m2
    return(n)
x = np.arange(-3, 3.1, 0.1)
y = np.arange(-4, 4.1, 0.1)
X, Y = np.meshgrid(x, y, indexing='ij')
nsq = modeexpsq(0, 1, 1, 0.5, X, Y)
plt.imshow(nsq)
```

## Free Energy Functional

### XPFC

The XPFC free energy functional consists the ideal energy contribution $F_{id}$ and the excess energy contribution $F_{ex}$:

$$
\begin{align}
F_{id}&=\int_V d\mathbf{r}\left[\frac{n^2(\mathbf{r})}{2}-\eta\frac{n^3(\mathbf{r})}{6}+\chi\frac{n^4(\mathbf{r})}{12}\right] \\
F_{ex}&=-\frac{1}{2}\int_V d\mathbf{r}\left[n(\mathbf{r})\int_V d\mathbf{r}'\left[C_2(\mathbf{r}-\mathbf{r}')n(\mathbf{r}')\right]\right]
\end{align}
$$

By adopting the convolution theorem, $C_2(\mathbf{r}-\mathbf{r}')n(\mathbf{r}')=\mathcal{F}^{-1}\left(\hat{C}_2(\mathbf{k})\cdot\hat{n}(\mathbf{k})\right)$, where $\hat{C}_2(\mathbf{k})\neq 0$ only for the length scale of the reciprocal family of planes.

* Substitute the amorphous phase's density $n_a(\mathbf{r})$, the triangle phase's density $n_t(\mathbf{r})$, and the square phase's density $n_s(\mathbf{r})$ to the ideal energy $F_{id}$, and divide by the unit volume/area to get the free energy density $f(n)$:

$$
\begin{align}
f_{id,a}(n) &= \frac{n^2}{2}-\eta \frac{n^3}{6}+ \chi \frac{n^4}{12} \\
f_{id,t}(n) &= \frac{n^2}{2}-\eta \frac{n^3}{6}+ \chi \frac{n^4}{12} + 3A_1^2(1-\eta n+\chi n^2) \nonumber \\
&-2A_1^3(\eta-2\chi n) + \frac{15A_1^4}{2}\chi \\
f_{id,s}(n) &= \frac{n^2}{2}-\eta \frac{n^3}{6}+ \chi \frac{n^4}{12} + 2(A_1^2+A_2^2)(1-\eta n+\chi n^2) \nonumber \\
&-4A_1^2A_2(\eta-2\chi n)+3(A_1^4+4A_1^2A_2^2+A_2^4)\chi
\end{align}
$$

it is not surprising to see that free energy density is **NOT** a function of lattice constant $a$, because the free energy functional has nothing to do with the lattice constant.

* For the excessed free energy part, the $\mathbf{k}=k_0=0$ planes (0th mode) provide a $n^2\hat{C}(k_0)$ term, the $\mathbf{k}=k_1$ planes provide a $A_1^2\hat{C}(k_1)$ term, and the $\mathbf{k}=k_2$ planes provide a $A_2^2\hat{C}(k_2)$ term. Such that 

$$
\begin{align}
f_{ex,a} &= -\frac{n^2}{2}\hat{C}(k_0) \\
f_{ex,t} &= -\frac{n^2}{2}\hat{C}(k_0) - 3A_1^2\hat{C}(k_1) \\
f_{ex,s} &= -\frac{n^2}{2}\hat{C}(k_0) - 2A_1^2\hat{C}(k_1) - 2A_2^2\hat{C}(k_2)
\end{align}
$$

### minimize the functional

The free energy density functionals should be minimized by tuning the amplitutdes.

$$
\begin{align}
f_{a}(n) &= \frac{n^2}{2}-\eta \frac{n^3}{6}+ \chi \frac{n^4}{12} -\frac{n^2}{2}\hat{C}(k_0)\\
f_{t}(n) &= \frac{n^2}{2}-\eta \frac{n^3}{6}+ \chi \frac{n^4}{12} + 3A_1^2(1-\eta n+\chi n^2) \nonumber \\
&-2A_1^3(\eta-2\chi n) + \frac{15A_1^4}{2}\chi -\frac{n^2}{2}\hat{C}(k_0) - 3A_1^2\hat{C}(k_1)\\
f_{s}(n) &= \frac{n^2}{2}-\eta \frac{n^3}{6}+ \chi \frac{n^4}{12} + 2(A_1^2+A_2^2)(1-\eta n+\chi n^2) \nonumber \\
&-4A_1^2A_2(\eta-2\chi n)+3(A_1^4+4A_1^2A_2^2+A_2^4)\chi \nonumber \\
&-\frac{n^2}{2}\hat{C}(k_0) - 2A_1^2\hat{C}(k_1) - 2A_2^2\hat{C}(k_2)
\end{align}
$$

A numerical minimization for $A$ yields:

```python
import numpy as np
import matplotlib.pyplot as plt

def faamp(nlst, eta, chi, C0):
    n = nlst
    f = n**2/2-eta*n**3/6+chi*n**4/12-n**2/2*C0 
    return(n, f)

def ftamp(nlst, eta, chi, C0, C1):
    amax = 1.0
    aunit = 0.0025
    alim = 0.1
    alst = np.arange(0, amax+aunit, aunit)
    [a1, n] = np.meshgrid(alst, nlst, indexing='ij')
    fmat = n**2/2-eta*n**3/6+chi*n**4/12 \
        + 3*a1**2*(1-eta*n+chi*n**2) \
        - 2*a1**3*(eta-2*chi*n) + 7.5*a1**4*chi \
        - 0.5*n**2*C0-3*a1**2*C1
    fmin = np.min(fmat, axis=0)
    amin = alst[np.argmin(fmat, axis=0)]
    nt = nlst[amin>alim]
    ft = fmin[amin>alim]
    return(nt, ft)

def fsamp(nlst, eta, chi, C0, C1, C2):
    amax = 1.0
    aunit = 0.0025
    alim = int(0.1/aunit)
    alst = np.arange(0, amax+aunit, aunit)
    [a1, a2, n] = np.meshgrid(alst, alst, nlst, indexing='ij')
    fmat = n**2/2-eta*n**3/6+chi*n**4/12 \
        + 2*(a1**2+a2**2)*(1-eta*n+chi*n**2) \
        - 4*a1**2*a2*(eta-2*eta*n) + 3*(a1**4+4*a1**2*a2**2+a2**4)*chi \
        - 0.5*n**2*C0 - 2*a1**2*C1 - 2*a2**2*C2
    fmin = np.min(fmat, axis=(0, 1))
    as_index = []
    for i in range(fmat.shape[2]):
        min_index = np.unravel_index(np.argmin(fmat[:, :, i]), fmat[:, :, i].shape)
        a1nonzero = min_index[0]>alim
        a2nonzero = min_index[1]>alim
        as_index.append(a1nonzero*a2nonzero)
    ns = nlst[as_index]
    fs = fmin[as_index]
    return(ns, fs)

nlst = np.arange(-0.3, 0.5, 0.01)
[na, fa] = faamp(nlst=nlst, eta=1, chi=1, C0=0.0)
[nt, ft] = ftamp(nlst=nlst, eta=1, chi=1, C0=0.0, C1=0.919)
[ns, fs] = fsamp(nlst=nlst, eta=1, chi=1, C0=0.0, C1=0.919, C2=0.887)
plt.figure()
plt.plot(na, fa, label='amourphous')
plt.plot(nt, ft, label='triangle')
plt.plot(ns, fs, label='square')
plt.grid()
plt.legend()
[nt, ft] = ftamp(nlst=nlst, eta=1, chi=1, C0=0.0, C1=0.969)
[ns, fs] = fsamp(nlst=nlst, eta=1, chi=1, C0=0.0, C1=0.969, C2=0.956)
plt.figure()
plt.plot(na, fa, label='amourphous')
plt.plot(nt, ft, label='triangle')
plt.plot(ns, fs, label='square')
plt.grid()
plt.legend()
```

#### temperature and specific length

The correlation parameter is where the temperature $\sigma$ and specific length $a$ goes in, according to *GREENWOOD M, PROVATAS N, ROTTLER J. Free Energy Functionals for Efficient Phase Field Crystal Modeling of Structural Phase Transformations[J/OL]. Physical Review Letters, 2010, 105(4): 045702. DOI:10.1103/PhysRevLett.105.045702*:

$$
\begin{align}
\hat{C}(k)=\max_{\text{for each peak}} \left\{e^{-\frac{\sigma^2k_i^2}{2\rho_i\beta_i}}e^{-\frac{(k-k_i)^2}{2\alpha_i^2}}\right\}
\end{align}
$$

However I believe there is a typo in $e^{-\frac{\sigma^2k_i^2}{2\rho_i\beta_i}}$, where the $k_i^2$ should be eliminated, by a numerical validation with fig. 1, it should be:

$$
\begin{align}
\hat{C}(k)=\max_{\text{for each peak}} \left\{e^{-\frac{\sigma^2}{2\rho_i\beta_i}}e^{-\frac{(k-k_i)^2}{2\alpha_i^2}}\right\}
\end{align}
$$

The specific length $a$ can be set based on the simulated system, in Greenwood's model, $a=1$, $k_1=\frac{2\pi}{a}\approx 6.28$, $k_2=\frac{2\sqrt{2}\pi}{a}\approx 8.88$, $\beta_1=\beta_2=4$, $\rho_1=1$, $\rho_2=1/\sqrt{2}$.

For example, at high (Ch*) and low (Cl*) temperature, the peaks' heights:

```python
import numpy as np
Ch1 = np.exp(-0.82**2/8)
Ch2 = np.exp(-0.82**2/5.657)
Cl1 = np.exp(-0.5**2/8)
Cl2 = np.exp(-0.5**2/5.657)
print(Ch1, Ch2, Cl1, Cl2)
```

considering the temperature $\sigma$ and the $C0$ term, the functions are defined as,

```python
import numpy as np
import matplotlib.pyplot as plt

def fasig(nlst, eta, chi, C0):
    n = nlst
    f = n**2/2-eta*n**3/6+chi*n**4/12-n**2/2*C0 
    return(n, f)

def ftsig(nlst, eta, chi, C0, sigma):
    C1 = np.exp(-sigma**2/8)
    amax = 1.0
    aunit = 0.0025
    alim = 0.1
    alst = np.arange(0, amax+aunit, aunit)
    [a1, n] = np.meshgrid(alst, nlst, indexing='ij')
    fmat = n**2/2-eta*n**3/6+chi*n**4/12 \
        + 3*a1**2*(1-eta*n+chi*n**2) \
        - 2*a1**3*(eta-2*chi*n) + 7.5*a1**4*chi \
        - 0.5*n**2*C0-3*a1**2*C1
    fmin = np.min(fmat, axis=0)
    amin = alst[np.argmin(fmat, axis=0)]
    nt = nlst[amin>alim]
    ft = fmin[amin>alim]
    return(nt, ft)

def fssig(nlst, eta, chi, C0, sigma):
    C1 = np.exp(-sigma**2/8)
    C2 = np.exp(-sigma**2/5.657)
    amax = 1.0
    aunit = 0.0025
    alim = int(0.1/aunit)
    alst = np.arange(0, amax+aunit, aunit)
    [a1, a2, n] = np.meshgrid(alst, alst, nlst, indexing='ij')
    fmat = n**2/2-eta*n**3/6+chi*n**4/12 \
        + 2*(a1**2+a2**2)*(1-eta*n+chi*n**2) \
        - 4*a1**2*a2*(eta-2*eta*n) + 3*(a1**4+4*a1**2*a2**2+a2**4)*chi \
        - 0.5*n**2*C0 - 2*a1**2*C1 - 2*a2**2*C2
    fmin = np.min(fmat, axis=(0, 1))
    as_index = []
    for i in range(fmat.shape[2]):
        min_index = np.unravel_index(np.argmin(fmat[:, :, i]), fmat[:, :, i].shape)
        a1nonzero = min_index[0]>alim
        a2nonzero = min_index[1]>alim
        as_index.append(a1nonzero*a2nonzero)
    ns = nlst[as_index]
    fs = fmin[as_index]
    return(ns, fs)

nlst = np.arange(-0.3, 0.5, 0.01)
dsig = 0.1
C0 = 0.0
eta = 1.0
chi = 1.0 
for sigma in np.arange(0.6, 1.0+dsig, dsig):
    [na, fa] = fasig(nlst=nlst, eta=eta, chi=chi, C0=C0)
    [nt, ft] = ftsig(nlst=nlst, eta=eta, chi=chi, C0=C0, sigma=sigma)
    [ns, fs] = fssig(nlst=nlst, eta=eta, chi=chi, C0=C0, sigma=sigma)
    plt.figure()
    plt.plot(na, fa, label='amourphous')
    plt.plot(nt, ft, label='triangle')
    plt.plot(ns, fs, label='square')
    plt.grid()
    plt.legend()
```
